```python
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt

pleiades = Table.read('C:/Users/dougl/Desktop/pleiades.csv', format = 'csv')

ngc188 = Table.read('C:/Users/dougl/Desktop/ngc188.csv', format = 'csv')
```

```python
print(pleiades[1].columns)
```

```python
print(ngc188[1].columns)
```

```python
pl_gmag = pleiades["phot_g_mean_mag"]
pl_bprp = pleiades["bp_rp"]
pl_ext = pleiades["a_g_val"]
pl_pmra,pl_pmdec = pleiades["pmra"], pleiades["pmdec"]
pl_ra, pl_dec = pleiades["ra"], pleiades["dec"]

ngc_gmag = ngc188["phot_g_mean_mag"]
ngc_bprp = ngc188["bp_rp"]
ngc_ext = ngc188["a_g_val"]
ngc_pmra,ngc_pmdec = ngc188["pmra"], ngc188["pmdec"]
ngc_ra, ngc_dec = ngc188["ra"], ngc188["dec"]
```

# Pleiades

```python
plt.scatter(pl_pmra,pl_pmdec, s=0.1)
plt.xlim(-50,50)
plt.ylim(-60,60)
plt.xlabel('$\mu_{a}$ [mas/yr]', fontsize=15)
plt.ylabel('$\mu_{\delta}$ [mas/yr]', fontsize=15)
```

```python
pl_d = 1000/(pleiades["parallax"])

pl_absgmag = pl_gmag - 5*np.log10(pl_d) + 5 + pl_ext

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,8))
ax[0].scatter(pl_bprp, pl_absgmag, s=.9, color='#666666')
ax[0].set_xlabel('BP-RP', fontsize=15)
ax[0].set_ylabel('Abs(G mag)', fontsize=15)
ax[0].set_ylim(ax[0].set_ylim(15,0)[::-1])
ax[0].set_title('Pleiades (1)', size=20)


ichoose1 = np.where((pl_pmra>18) & (pl_pmra<22) & (pl_pmdec>-58) & (pl_pmdec<-39) 
               & (pleiades['parallax'] > 5.5) & (pleiades['parallax'] < 8))

ax[1].scatter(pl_bprp, pl_absgmag, s=.4, color='#666666', alpha=0.5)
ax[1].scatter(pl_bprp[ichoose1], pl_absgmag[ichoose1], s=3, color='red', marker='o')


ax[1].set_xlabel('BP-RP', fontsize=15)
ax[1].set_ylabel('Abs(G mag)', fontsize=15)
ax[1].set_ylim(ax[1].set_ylim(15,0)[::-1])

#plt.savefig('cmd_ple.png', type='png')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,8))
ax[0].scatter(pl_bprp, pl_gmag, s=.9, color='#666666')
ax[0].set_xlabel('BP-RP', fontsize=15)
ax[0].set_ylabel('G mag', fontsize=15)
ax[0].set_ylim(ax[0].set_ylim(6,21)[::-1])
ax[0].set_title('Pleiades (2)', size=16)


ichoose2 = np.where((pl_pmra>18) & (pl_pmra<22) & (pl_pmdec>-58) & (pl_pmdec<-39) 
               & (pleiades['parallax'] > 5.5) & (pleiades['parallax'] < 8))

ax[1].scatter(pl_bprp, pl_gmag, s=.4, color='#666666', alpha=0.5)
ax[1].scatter(pl_bprp[ichoose2], pl_gmag[ichoose2], s=3, color='red', marker='o')


ax[1].set_xlabel('BP-RP', fontsize=15)
ax[1].set_ylabel('G mag', fontsize=15)
ax[1].set_ylim(ax[1].set_ylim(6,21)[::-1])
```
